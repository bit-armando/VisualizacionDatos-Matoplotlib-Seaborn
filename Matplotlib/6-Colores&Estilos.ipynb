```python
import matplotlib.pyplot as plt
import numpy as np

th = np.linspace(0, 2*np.pi, 128)
```

Visualizamos los estilos por defecto que tiene matplotlib

```python
print(plt.style.available)
```

```python
def demo(sty):
    plt.style.use(sty)
    
    fig, ax = plt.subplots(figsize=(5,5))
    ax.set_title('style: {!r}'.format(sty), color='C0')
    
    ax.plot(th, np.cos(th), label='C1')
    ax.plot(th, np.sin(th), label='C2')
    ax.legend()
    plt.show()
```

```python
demo('tableau-colorblind10')
```
