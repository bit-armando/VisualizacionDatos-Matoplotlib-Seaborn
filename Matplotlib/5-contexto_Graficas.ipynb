Para darle contexto a las graficas se necesitan de usar titulos, leyendas, tamaño o tamaño.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 5, 11)
y = np.sin(x)
```

## Titulos

```python
fig, axes = plt.subplots(1, 2, figsize =(10,5)) # figsize cambia el tamaño del lienzo
axes[0].set_title('Relacion X - Y') # Agrega un titulo
axes[0].set_xlabel('X') # Agrega un label a eje X
axes[0].set_ylabel('Y') # Agrega un label a eje Y   
axes[0].plot(x, y, label = '$sin(x)$')
axes[0].legend() # Agrega la leyenda que definio al plot
axes[1].set_title('Relacion Y - X')
axes[1].plot(y, x, 'r')
plt.show()
```
