```python
import numpy as np
import matplotlib.pyplot as plt
```

Creamos nuestras variables

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

Graficamos en funcion de y

```python
plt.plot(x, y)
plt.show()
```

### Modificar graficas usando pyplot

|Character| Color |
|---------|-------|
|'b'      |blue   |
|'g'      |green  |
|'r'      |red    |
|'m'      |magenta|

Hay mas parametros como la forma de la linea y como se contecta

```python
plt.plot(x, y, 'mD-')
plt.show()
```

## Crear un histograma

```python
plt.hist(y)
plt.show()
```

## Crear una grafica de pastel

```python
plt.pie(y)
plt.show()
```

## Conocer correlacion entre variables

```python
plt.scatter(x, y)
plt.show()
```

## Distribucion de datos

```python
plt.boxplot(x)
plt.show()
```
