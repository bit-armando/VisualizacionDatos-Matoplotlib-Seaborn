Subplot permite crear una grafica dentro de una grafica.
Esto es lo que hace atravez de una matriz de graficos y se puede acceder a ellos a travez de indices:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 5, 11)
y = x ** 2
```

## Parametros en subplot

|Filas|Columnas|Index|
|-----|--------|-----|

```python
plt.subplot(1, 2, 1) # Grafico 1
plt.plot(x, y, 'r--')

plt.subplot(1, 2, 2) # Grafico 2
plt.pie(y)
plt.show()
```

Agregando diferentes plot al grafico

```python
plt.subplot(1, 2, 1) # Grafico 1
plt.plot(x, y, 'r--')
plt.plot(y, x, 'mo--')

plt.subplot(1, 2, 2) # Grafico 2
plt.pie(y)
plt.show()
```
