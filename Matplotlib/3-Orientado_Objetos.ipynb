Hay diferentes maneras de crear un grafico en matplotlib como pyplot, pero si usamos esta manera nos seria muy complicado personalisar nuestros graficos y entrar a ciertos parametros

Un objeto define una figura, esta es un lienzo en donde en el cual podemos introducir diferentes `graficas(axes)` de las cuales poseen sus propios `ejes(axis)`

<img src='../img/Estructura.png'>

## Diferencias entre pyplot y Object Oriented

|Pyplot          | Object Oriented                   |
|----------------|-----------------------------------|
|Rapido          | Mayor personalizacion             |
|Facil           | Mas codigo                        |
|Una sola figura | Mas amigable a multiples diagramas|

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 5, 11)
y = x ** 2
```

### Parametros de axes

|Pos.EjeX|Pos.EjeY|SizeY|SizeX|
|--------|--------|-----|-----|

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.9])
axes.plot(x,y, 'b')

axes2 = fig.add_axes([0.17, 0.55, 0.4, 0.3])
axes2.plot(y,x, 'r:')
axes2.set_facecolor('grey')
plt.show()
```
