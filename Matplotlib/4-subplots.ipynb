Con `subplots` se puede trabajar en un **arreglo de graficas** a las cuales se acceden a travez de indices

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 5, 11)
y = x ** 2
```

### Crear una grafica tipo fig

```python
fig, axes = plt.subplots()
axes.plot(x, y)
plt.show()
```

Varias figuras

```python
fig, axes = plt.subplots(nrows= 1, ncols=2)
axes[0].plot(x, y)
axes[1].plot(y, x, 'r')
plt.show()
```
