Seaborn ofrece graficas u opciones que permiten trabajar con distribuciones de datos numericos

```python
import seaborn as sns
tips = sns.load_dataset('tips')
tips.head()
```

## Histograma

```python
sns.histplot(data=tips, x='tip');
```

```python
sns.histplot(data=tips, x='tip', hue='sex');
```

## Diagrama de densidad

```python
sns.kdeplot(data=tips, x='tip', hue='sex');
```

```python
sns.kdeplot(data=tips, x='tip', hue='sex', shade=True);
```

## Uso de `.displot()`

```python
sns.displot(data=tips, x='tip', hue='sex', kind='ecdf');
```
