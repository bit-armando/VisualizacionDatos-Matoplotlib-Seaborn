# Set_theme
Es un metodo que permite configurar el estilo, fuente y color de las graficas 

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

set_theme: https://seaborn.pydata.org/generated/seaborn.set_theme.html#seaborn.set_theme

Paletas de colores: https://seaborn.pydata.org/generated/seaborn.color_palette.html#seaborn.color_palette

```python
sns.set_theme()
sns.barplot(x=['A','B','C'], y=[1,3,2])
plt.show()
```
